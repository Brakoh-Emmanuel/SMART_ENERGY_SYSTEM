```python
"""
Smart Device Management System
Course: EL 162 / 234 Object Oriented Programming
Author: Student Assignment Submission
"""

class SmartDevice:
    """Parent class representing a generic smart device."""

    def __init__(self, name: str, device_id: str):
        # Public Attribute
        self.name = name

        # Private Attributes (Encapsulation via double underscore)
        if not device_id.strip():
            raise ValueError("Device ID cannot be empty.")
        self.__device_id = device_id
        self.__power_status = False  # Initially off

    # Getter and Setter for device_id
    @property
    def device_id(self) -> str:
        return self.__device_id

    # Getter and Setter for power_status
    @property
    def power_status(self) -> bool:
        return self.__power_status

    def turn_on(self):
        """Turns the smart device on."""
        self.__power_status = True
        print(f"[{self.name}] has been turned ON.")

    def turn_off(self):
        """Turns the smart device off."""
        self.__power_status = False
        print(f"[{self.name}] has been turned OFF.")

    def display_info(self):
        """Displays the baseline status information of the device."""
        status = "ON" if self.__power_status else "OFF"
        print(f"Device Name: {self.name}")
        print(f"Device ID:   {self.__device_id}")
        print(f"Power:       {status}")


class TemperatureSensor(SmartDevice):
    """Child class representing a smart temperature sensor."""

    def __init__(self, name: str, device_id: str, temperature: float = 24.0):
        # Call the parent class constructor using super()
        super().__init__(name, device_id)
        self.temperature = temperature

    def read_temperature(self):
        """Simulates reading the sensor temperature if power is on."""
        if not self.power_status:
            print(f"Error: {self.name} is offline. Please turn it on first.")
            return
        print(f"[{self.name}] Current Temperature: {self.temperature}°C")


class SmartLight(SmartDevice):
    """Child class representing an adjustable smart light bulb."""

    def __init__(self, name: str, device_id: str, brightness: int = 50):
        super().__init__(name, device_id)
        # Enforce validation bounds immediately upon initialization
        if not (0 <= brightness <= 100):
            raise ValueError("Brightness must be between 0 and 100.")
        self.__brightness = brightness

    @property
    def brightness(self) -> int:
        return self.__brightness

    @brightness.setter
    def brightness(self, value: int):
        if 0 <= value <= 100:
            self.__brightness = value
        else:
            print("Error: Brightness value must be strictly between 0 and 100.")

    def increase_brightness(self, amount: int = 10):
        """Increases brightness by a set increment up to 100."""
        if not self.power_status:
            print(f"Error: {self.name} is offline.")
            return
        self.brightness = min(100, self.__brightness + amount)
        print(f"[{self.name}] Brightness increased to: {self.__brightness}%")

    def decrease_brightness(self, amount: int = 10):
        """Decreases brightness by a set decrement down to 0."""
        if not self.power_status:
            print(f"Error: {self.name} is offline.")
            return
        self.brightness = max(0, self.__brightness - amount)
        print(f"[{self.name}] Brightness decreased to: {self.__brightness}%")

    def display_info(self):
        """Overrides parent display_info to add specific metrics."""
        super().display_info()
        print(f"Brightness:  {self.__brightness}%")


class SecurityCamera(SmartDevice):
    """Child class representing a smart security streaming camera."""

    def __init__(self, name: str, device_id: str):
        super().__init__(name, device_id)
        self.recording_status = False

    def start_recording(self):
        """Starts live recording if device is online."""
        if not self.power_status:
            print(f"Error: {self.name} is offline.")
            return
        self.recording_status = True
        print(f"[{self.name}] Video stream is now RECORDING.")

    def stop_recording(self):
        """Stops live recording."""
        self.recording_status = False
        print(f"[{self.name}] Video stream has STOPPED recording.")

    def display_info(self):
        """Overrides parent display_info to add camera specific metrics."""
        super().display_info()
        rec_txt = "RECORDING" if self.recording_status else "IDLE"
        print(f"Rec Status:  {rec_txt}")


# --- Interactive Menu System Interface ---
def main():
    # Instantiate instances to demonstrate all mandatory baseline methods
    sensor = TemperatureSensor("Living Room Sensor", "TEMP-001", 22.5)
    light = SmartLight("Bedroom Main Light", "LGT-404", 75)
    camera = SecurityCamera("Front Porch Camera", "CAM-999")

    # Group them systematically into an iterable tracking registry
    devices = [sensor, light, camera]

    while True:
        print("\n" + "="*35)
        print(" SMART DEVICE MANAGEMENT SYSTEM")
        print("="*35)
        print("1. Display Device Information")
        print("2. Turn Device On")
        print("3. Turn Device Off")
        print("4. Read Temperature")
        print("5. Adjust Brightness")
        print("6. Start Recording")
        print("7. Exit")
        print("="*35)

        choice = input("Select an option (1-7): ").strip()

        if choice == "1":
            print("\n--- Current Device Statuses ---")
            for dev in devices:
                dev.display_info()
                print("-" * 20)

        elif choice in ["2", "3"]:
            print("\nSelect a device:")
            for idx, dev in enumerate(devices, 1):
                print(f"{idx}. {dev.name} ({dev.device_id})")

            try:
                target_idx = int(input("Enter device number: ")) - 1
                if 0 <= target_idx < len(devices):
                    if choice == "2":
                        devices[target_idx].turn_on()
                    else:
                        devices[target_idx].turn_off()
                else:
                    print("Invalid device selection.")
            except ValueError:
                print("Please enter a valid numeric choice.")

        elif choice == "4":
            sensor.read_temperature()

        elif choice == "5":
            if not light.power_status:
                print(f"Error: Turn on {light.name} first before adjusting brightness.")
                continue
            print(f"Current brightness: {light.brightness}%")
            print("1. Increase Brightness (+10%)\n2. Decrease Brightness (-10%)")
            sub_choice = input("Selection: ").strip()
            if sub_choice == "1":
                light.increase_brightness()
            elif sub_choice == "2":
                light.decrease_brightness()
            else:
                print("Invalid operational request.")

        elif choice == "6":
            if not camera.power_status:
                print(f"Error: Turn on {camera.name} first before adjusting camera functions.")
                continue
            print("1. Start Live Feed Recording\n2. Stop Recording")
            sub_choice = input("Selection: ").strip()
            if sub_choice == "1":
                camera.start_recording()
            elif sub_choice == "2":
                camera.stop_recording()
            else:
                print("Invalid operational request.")

        elif choice == "7":
            print("\nShutting down system safely. Goodbye!")
            break
        else:
            print("Invalid input selection. Please enter a valid menu option between 1 and 7.")

if __name__ == "__main__":
    main()
```
